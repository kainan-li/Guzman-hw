```python
import numpy as np
import pandas as pd
import random
import math
```

```python
df = pd.read_excel('Assignment 4 - dataset.xlsx')
df['Date (America/Chicago)'] = pd.to_datetime(df['Date (America/Chicago)'])
df.set_index('Date (America/Chicago)',inplace = True)
df = df.resample('D').interpolate()[::7]
```

```python
randomlist = []
for i in range(44):
    n = random.uniform(23,28)
    randomlist.append(n)
df['predicted_product_1_mkt_price_on_execution_date'] = randomlist    
```

```python
randomlist = []
for i in range(44):
    n = random.uniform(2,3.5)
    randomlist.append(n)
df['predicted_product_2_mkt_price_on_execution_date'] = randomlist
```

```python
def fun(row):
    profit_product_1 = -5000 * (row['predicted_product_1_mkt_price_on_execution_date'] - 24.62)
    if profit_product_1 >= 10000:
        return 'No action'
    elif row['predicted_product_2_mkt_price_on_execution_date'] > row['Product 2']:
        long = math.ceil((10000 - profit_product_1) / (row['predicted_product_2_mkt_price_on_execution_date'] - row['Product 2']))
        return long
    else:
        return 'No action'
df['position for product 2'] = df.apply(lambda x:fun(x),axis = 1)    
```
